```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
# Read dataset
df = pd.read_csv('Results_21Mar2022.csv')
# display the first 5 rows of the dataset (0-4 rows)
df.head()
```

```python
# display information about the dataset
df.info()
```

```python
# display information about the dataset
df.describe()
```

```python
# get unique food type, gender and age 
diet_groups = df['diet_group'].unique()
sex_groups = df['sex'].unique()
age_groups = df['age_group'].unique()

print(f"unique food type: {diet_groups}")
print(f"unique gender: {sex_groups}")
print(f"unique age: {age_groups}")
```

```python
# choose the metrics to be plotted
metrics = [
    'mean_ghgs',        # Greenhouse Gas Emissions
    'mean_land',        # Land Use
    'mean_watscar',     # Water Scarcity
    'mean_eut',         # Eutrophication Potential
    'mean_ghgs_ch4',    # Greenhouse Gas Emissions (CH4)
    'mean_ghgs_n2o',    # Greenhouse Gas Emissions (N2O)
    'mean_bio',         # Biodiversity
    'mean_watuse',      # Water Use
    'mean_acid',        # Acidification Potential
]
```

## 1. Radar Chart & Parallel Coordinates Plot

```python
# set colors for each food type
foodtype_color_dict = {
    'vegan': 'lightgreen',      
    'veggie': 'green', 
    'fish': 'blue',
    'meat': 'orange',         
    'meat50': 'red',      
    'meat100': 'purple'    
}
```

```python
# standardize the metrics for diet groups
grouped_diet = df.groupby('diet_group')[metrics].mean().reset_index()
scaler = MinMaxScaler()
scaled_vals = scaler.fit_transform(grouped_diet[metrics])

# create a new DataFrame to store the scaled values
grouped_scaled = grouped_diet.copy()
for i in range(len(metrics)):
    col = metrics[i]
    grouped_scaled[col] = scaled_vals[:, i]

# create a new DataFrame with the scaled values
grouped_diet_scaled = grouped_diet.copy()
grouped_diet_scaled[metrics] = scaled_vals

# prepare subplot
fig = plt.figure(figsize=(16, 8))

# 1) Radar chart subplot
ax1 = fig.add_subplot(1, 2, 1, polar=True)
# angles for radar
N = len(metrics)
angles = []
for i in range(N):
    angles.append(2 * np.pi * i / N)
angles.append(angles[0])

# plot each diet group
for idx in range(grouped_diet_scaled.shape[0]):
    grp = grouped_diet_scaled.loc[idx, 'diet_group']
    if grp not in foodtype_color_dict:
        continue
    # calculate the value for each metric
    values = []
    for col in metrics:
        values.append(grouped_diet_scaled.loc[idx, col])
    values.append(values[0])
    c = foodtype_color_dict[grp]
    ax1.plot(angles, values, linewidth=2, label=grp, color=c)
    ax1.fill(angles, values, alpha=0.1, color=c)

ax1.set_xticks(angles[:-1])
ax1.set_xticklabels(metrics, fontsize=9)
ax1.set_title('Radar of Extended Metrics by Diet', pad=20)
ax1.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# 2) Parallel Coordinates Plot subplot
ax2 = fig.add_subplot(1, 2, 2)
palette = []
for grp in grouped_diet_scaled['diet_group']:
    if grp in foodtype_color_dict:
        palette.append(foodtype_color_dict[grp])

pd.plotting.parallel_coordinates(
    grouped_diet_scaled,
    'diet_group',
    ax=ax2,
    color=palette,
    alpha=0.7
)
ax2.set_xticks(range(len(metrics)))
ax2.set_xticklabels(metrics, rotation=45, ha='right', fontsize=9)
ax2.set_title('Parallel Coordinates of Extended Metrics by Diet')
ax2.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))

plt.tight_layout()
plt.show()
```

######
Perspective:

The Radar and Parallel Coordinates Plots present that 'vegan' diets consistently yield the lowest normalized environmental impacts across all the 9 metrics, followed closely by 'veggie' diets; 'meat50' diet reduces impacts compared to typical omnivorous 'meat' diet or 'fish' diet, while the 'meat100' diet produces the highest impacts in every category. More specifically, The largest impact occur in water use and The second largest impact occur in biodiversity. 

Overall, the clear rank in environmental performance is: vegan > veggie > meat50 > fish > meat > meat100, demonstrating that reducing or eliminating meat could make the greatest environmental benefits.

## 2. Box Plot

```python
# set colors for each age group
# '20-29' '30-39' '40-49' '50-59' '60-69' '70-79'
agegroup_color_dict = {
    '20-29': 'lightgreen',      
    '30-39': 'green', 
    '40-49': 'lightblue',
    '50-59': 'pink',        
    '60-69': 'red',      
    '70-79': 'orange',    
}
```

```python
# standardize the metrics for each age group
scaler = MinMaxScaler()
df_scaled = df.copy()
for col in metrics:
    vals = df_scaled[col].values.reshape(-1, 1)
    df_scaled[col] = scaler.fit_transform(vals)[:, 0]
df_scaled['composite_score'] = df_scaled[metrics].sum(axis=1)

# map age groups to colors
age_groups = sorted(df_scaled['age_group'].unique())
colors = list(agegroup_color_dict.values())
palette = {age: colors[i % len(colors)] for i, age in enumerate(age_groups)}


sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))


ax = sns.boxplot(
    x='age_group',
    y='composite_score',
    hue='age_group',
    data=df_scaled,
    palette=palette,
    showfliers=True,
    notch=False,
    dodge=False,
    legend=True
)

ax.set_title('Composite Environmental Score by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Composite Score')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

######
Perspective:

The Box Plot comparison of composite environmental scores by age group indicates that the '40–49' age group has the lowest median score which means this is the most environmentally friendly dietary patterns. However, the '70–79' age group has the highest median, suggesting the least sustainable diets. Although all age bands display similar interquartile ranges, the youngest (20–29) and oldest (70–79) groups show slightly greater variability, and each contains extreme high-score outliers. 

Generally, the median score trends upward from middle to old age with the 40–49 group emerging as the clear low-impact leader.

## 3. Heat Map

```python
title_map = {
    'mean_ghgs':     'GHG',
    'mean_land':     'Land Use',
    'mean_watscar':  'Water Scarcity',
    'mean_eut':      'Eutrophication',
    'mean_ghgs_ch4': 'CH4',
    'mean_ghgs_n2o': 'N2O',
    'mean_bio':      'Biodiversity',
    'mean_watuse':   'Water Use',
    'mean_acid':     'Acidification'
}

# create a 3×3 subplot grid
fig, axes = plt.subplots(
    nrows=3, ncols=3,
    figsize=(18, 18),
    constrained_layout=True
)

# 4. loop over metrics and axes
for idx, metric in enumerate(metrics):
    ax = axes.flatten()[idx]
    # pivot table: diet_group × age_group
    pivot = df.pivot_table(
        index='diet_group',
        columns='age_group',
        values=metric,
        aggfunc='mean'
    )
    
    sns.heatmap(
        pivot,
        annot=True,
        fmt='.2f',
        cmap='coolwarm',
        cbar=False, 
        ax=ax
    )
    # set sub titles and labels
    ax.set_title(title_map[metric], fontsize=14)
    ax.set_xlabel('Age Group')
    ax.set_ylabel('Diet Group')

# set title
fig.suptitle(
    'Average Environmental Metrics by Diet and Age Groups',
    fontsize=18,
    y=1.02
)

plt.show()
```

######
Perspective:

The heatmap panel clearly illustrates that across every metric—greenhouse gas emissions, land use, water scarcity, eutrophication potential, CH₄ and N₂O emissions, biodiversity impact, water use, and acidification potential the 'meat100' diet has the highest average impact in all age groups, followed by the 'meat' diet. In contrast, the 'vegan' diet consistently register the lowest values, with the 'veggie' diet slightly higher, both far below any meat‐inclusive diet. 

More specifically, the '70-79' age group with 'meat100' diet has the highest impact on the aspect of greenhouse gas emissions, CH₄ and N₂O emissions, land use, biodiversity and acidification. Apart from that,  the youngest age group (20-29) with 'meat100' diet has the highest impact on the aspect of water scarcity, eutrophication and water use.  

In conclusion, eliminating or sharply reducing meat intake offers the single greatest reduction in environmental impact for all age groups.

## Parallel Coordinates Plot

```python
# set colors for gender
gender_color_dict = {'male': 'skyblue', 'female': 'lightpink'}
```

```python
# standardize the metrics
scaler = MinMaxScaler()
df_scaled = df.copy()
for col in metrics:
    vals = df_scaled[col].values.reshape(-1, 1)
    df_scaled[col] = scaler.fit_transform(vals)[:, 0]

# calculate the composite score  
df_scaled['composite_score'] = df_scaled[metrics].sum(axis=1)

plt.figure(figsize=(8, 6))
ax = sns.violinplot(
    x='sex',
    y='composite_score',
    hue='sex',
    data=df_scaled,
    palette=gender_color_dict,
    inner='box', 
    dodge=False,
    legend= True
)

ax.set_title('Composite Environmental Score by Gender', fontsize=14)
ax.set_xlabel('Gender')
ax.set_ylabel('Composite Score')

plt.tight_layout()
plt.show()
```

######
Perspective:

The violin plots show that females tend to have slightly lower composite environmental scores than males: the median for females (around 1.4) is lower than the median for males (around 1.6), and the distribution for females is much more concentrated, with a noticeable peak at the lower scores. In contrast, males have a wider distribution, especially in the upper tail, suggesting more high-impact outliers. 

Despite this overlap, the overall offset and tighter inter-quartile range for females suggests that, on average, females have more environmentally friendly and consistent eating patterns than males.
